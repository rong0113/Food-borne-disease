```python
import re
import os #Import for setting working dirrectory
import requests #Import for downloading data from database
from tqdm import tqdm
import sys
import numpy as np
import warnings
warnings.filterwarnings("ignore")
from tqdm.notebook import tqdm
url_base = "https://www.ncei.noaa.gov/data/global-summary-of-the-day/access/" #Daily data path 
for i in tqdm(range(len(df_ny))):
    wban_unmodified = df_ny.loc[i, "WBAN"]
    if 0 < wban_unmodified < 10000: 
        df_ny.loc[i, "WBAN"] = str(wban_unmodified).zfill(5)
    usaf = df_ny.loc[i, "USAF"]
    #print(usaf) 
    wban = df_ny.loc[i, "WBAN"]
    filename_noaa = str(usaf) + str(wban) + ".csv"
    for yr in range(2011,2020):
        url_loop_daily = url_base + str(yr) + "/" + filename_noaa
                    #print(url_loop_daily)

                    #Identifies file on website
        csv_current_daily = requests.get(url_loop_daily, allow_redirects = True)

                    #Sets the file name of the local imported file
        filename_local_daily =  str(usaf) + str(wban) + "-" + str(yr) + "DAILY.csv" 
        os.chdir('/Users/evelyn/Desktop/Frequency/NYM') #Sets path for daily data csv deposit  
        open(filename_local_daily, "wb").write(csv_current_daily.content)    
```

```python
startdate = datetime.datetime.strptime(str('2011/01/01'), "%Y/%m/%d")
enddate = datetime.datetime.strptime(str('2019/12/31'), "%Y/%m/%d")
d = startdate
delta = datetime.timedelta(days=1)
path = '/Users/evelyn/Desktop/Frequency/NYM/'
import numpy as np
index = 0
while d <= endtime:
    yr = d.year
    temp = []; slp = []; stp = []; wdsp=[]; prcp=[]; mxspd=[]; dewp=[];
    for i in range(len(df_ny)):
        usaf = df_ny.loc[i, "USAF"] 
        wban = df_ny.loc[i, "WBAN"]
        filename = str(usaf) + str(wban) + "-" + str(yr) + "DAILY.csv"
        df = pd.read_csv(path+filename)
        if(sum(1 for line in open(path+filename)) < 15):
            continue 
        else:
            if(df.DATE.str.contains(str(d)[:10]).any()):
                m = df[df.DATE == str(d)[:10]].index.tolist()[0]
                if(df.loc[m, "TEMP"] < 9999):
                    temp.append(df.loc[m, "TEMP"])
                if(df.loc[m, "DEWP"] < 9999):
                    dewp.append(df.loc[m, "DEWP"])
                if(df.loc[m, "SLP"] < 9999):
                    slp.append(df.loc[m, "SLP"])
                if( 100<df.loc[m, "STP"] < 9999 and df.loc[m, "STP_ATTRIBUTES"]>0):
                    stp.append(df.loc[m, "STP"])
                elif(100>df.loc[m, "STP"]and df.loc[m, "STP_ATTRIBUTES"]>0):
                    stp.append(df.loc[m, "STP"]+1000)  
                if(df.loc[m, "WDSP"] < 999):
                    wdsp.append(df.loc[m, "WDSP"])
                if(df.loc[m, "MXSPD"] < 998):
                    mxspd.append(df.loc[m, "MXSPD"])
                if(df.loc[m, "PRCP"] < 99):
                    prcp.append(df.loc[m, "PRCP"])
        if(len(temp) > 0):
            fre_NY.loc[index, "TEMP"] = np.mean(temp)
        if(len(slp) > 0):
            fre_NY.loc[index, "SLP"] = np.mean(slp)
        if(len(stp) > 0):
            fre_NY.loc[index, "STP"] = np.mean(stp)
        if(len(wdsp) > 0):
            fre_NY.loc[index, "WDSP"] = np.mean(wdsp)
        if(len(prcp) > 0):
            fre_NY.loc[index, "PRCP"] = np.mean(prcp)
        if(len(dewp) > 0):
            fre_NY.loc[index, "DEWP"] = np.mean(dewp)
        if(len(mxspd) > 0):
            fre_NY.loc[index, "MXSPD"] = np.mean(mxspd)
        
    d = d + delta
    index = index + 1
```

```python
fre_NY.isnull().sum()
```

```python
import math
for i in range(len(fre_NY)):
    t_c = (fre_NY.loc[i, "TEMP"] - 32)/1.8 
    d_c = (fre_NY.loc[i, "DEWP"]-32)/1.8
    rh = 100 * (math.exp(17.27 * d_c/(237.7 + d_c))/math.exp(17.27 * t_c/(237.7 + t_c)))
    fre_NY.loc[i, "RH"] = rh

for i in range(len(fre_NY)):
    fre_NY.loc[i, "SLP-STP"] =  fre_NY.loc[i, "SLP"] - fre_NY.loc[i, "STP"]

for i in range(len(fre_NY)):
    fre_NY.loc[i, "MSXPD-WDSP"] =  fre_NY.loc[i, "MXSPD"] - fre_NY.loc[i, "WDSP"]
```
