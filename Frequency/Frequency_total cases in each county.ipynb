```python
# import data
import pandas as pd
frequency = pd.read_csv('/Users/evelyn/Desktop/weather information/Outbreak Frequency.csv')
# Sets columns Total cases of salmonella and norovirus and initialize it as 0
frequency['Salm Total cases'] = 0
frequency['Noro Total cases'] = 0
```

```python
#import salm data and calculate the total cases spanning from 2010 to 2019

salm = pd.read_csv('/Users/evelyn/Desktop/weather information/SALM2010-2019.csv')

for i in range(len(salm)):
    a = 0
    for j in range(0,19):
        if not(pd.isna(salm.loc[i, "ExposureCounty"+str(j+1)])):
            a = a + 1
        else:
            break
    for z in range(0,19):
        if not(pd.isna(salm.loc[i, "ExposureCounty"+str(z+1)])):
            df = frequency.loc[frequency.USPS.str.contains(str(salm.loc[i,"Exposurestate"])) & frequency.NAME1.str.contains(str(salm.loc[i,"ExposureCounty"+str(z+1)]))]
            index = df.index.tolist()[0]  
            frequency.loc[index, "Salm Total cases"] = frequency.loc[index, "Salm Total cases"] + (salm.loc[i, "TotalCases"])/a

        
        
```

```python
#import noro data and calculate the total cases spanning from 2010 to 2019

noro = pd.read_csv('/Users/evelyn/Desktop/weather information/NORO2010-2019.csv')

for i in range(len(noro)):
    a = 0
    for j in range(0,11):
        if not(pd.isna(noro.loc[i, "ExposureCounty"+str(j+1)])):
            a = a + 1
        else:
            break
    for z in range(0,11):
        if not(pd.isna(noro.loc[i, "ExposureCounty"+str(z+1)])):
            df = frequency.loc[frequency.USPS.str.contains(str(noro.loc[i,"Exposurestate"])) & frequency.NAME1.str.contains(str(noro.loc[i,"ExposureCounty"+str(z+1)]))]
            index = df.index.tolist()[0]  
            frequency.loc[index, "Noro Total cases"] = frequency.loc[index, "Salm Total cases"] + (noro.loc[i, "TotalCases"])/a
        
        
```

```python
# Caculate the SALM total num of outbreaks in each county 
for i in range(len(frequency)):
    a = 0
    df = salm[salm.Exposurestate.str.contains(str(frequency.loc[i,"USPS"]))]
    if(len(df) == 0):
        frequency.loc[i,"SALM"] = 0
    else:
        for j in range(0,19):
            a = a + len(df[df["ExposureCounty"+str(j+1)]==str(frequency.loc[i,"NAME1"])])
        frequency.loc[i, "SALM"] = int(a)
    
```

```python
# Caculate the NORO total num of outbreaks in each county 
for i in range(len(frequency)):
    a = 0
    df = noro[noro.Exposurestate.str.contains(str(frequency.loc[i,"USPS"]))]
    if(len(df) == 0):
        frequency.loc[i,"NORO"] = 0
    else:
        for j in range(0,11):
            a = a + len(df[df["ExposureCounty"+str(j+1)]==str(frequency.loc[i,"NAME1"])])
        frequency.loc[i, "NORO"] = int(a)
    
```

```python
frequency['SALM'] = frequency['SALM'].astype(int)
frequency['NORO'] = frequency['NORO'].astype(int)
frequency['Salm Total cases'] = frequency['Salm Total cases'].astype(int)
frequency['Noro Total cases'] = frequency['Noro Total cases'].astype(int)
frequency = frequency.drop(['SALM1'], axis = 1)
frequency
```
